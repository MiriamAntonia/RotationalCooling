```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.special import erf

# === ERROR FUNCTION MODEL ===
def erf_edge(x, A, x0, sigma, B):
    return A * 0.5 * (1 + erf((x - x0) / (np.sqrt(2) * sigma))) + B
```

```python
# === USER INPUT ===
file_air = "FirstPurging/BeforePurging/2770nm_04_0.51nm.txt"

# wavelength range (in nm)
xmin = 2640.0
xmax = 2643.85

# === LOAD DATA ===
# assumes delimiter is ";" like in your screenshot
data_air = np.loadtxt(file_air, delimiter=";")

wl_air, int_air = data_air[:, 0], data_air[:, 1]

# === FILTER RANGE ===
mask_air = (wl_air >= xmin) & (wl_air <= xmax)

wl_air_f, int_air_f = wl_air[mask_air], int_air[mask_air]

# === INITIAL GUESSES ===
A_guess = np.max(int_air_f) - np.min(int_air_f)
B_guess = np.min(int_air_f)
x0_guess = wl_air_f[np.argmax(np.gradient(int_air_f))]  # edge position
sigma_guess = 0.05  # nm (adjust if needed)

p0 = [A_guess, x0_guess, sigma_guess, B_guess]

# === FIT ===
popt, pcov = curve_fit(erf_edge, wl_air_f, int_air_f, p0=p0)

A_fit, x0_fit, sigma_fit, B_fit = popt

# === CONVERT TO FWHM ===
twosigma = 2* sigma_fit
FWHM = 2 * np.sqrt(2 * np.log(2)) * sigma_fit

print(f"Edge position λ0 = {x0_fit:.4f} nm")
print(f"2 Sigma = {twosigma:.4f} nm")
print(f"FWHM (resolution) = {FWHM:.4f} nm")

# === PLOT ===
x_fit = np.linspace(xmin, xmax, 500)
y_fit = erf_edge(x_fit, *popt)

plt.plot(wl_air_f, int_air_f, label="Data", linewidth=0.8)
plt.plot(x_fit, y_fit, '--', label="Erf fit", linewidth=2)

plt.xlabel("Wavelength (nm)", fontsize=13)
plt.ylabel("Intensity", fontsize=13)
plt.legend()
plt.xlim(xmin, xmax)
plt.tight_layout()
plt.show()
```
